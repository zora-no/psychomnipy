```python
def sem (data):
    return np.std(data, ddof=1, axis=1) / np.sqrt(np.size(data))
```

```python
plt.figure(figsize=(10,10))

plt.plot(sem(ohne), color='b');
plt.plot(sem(innerhalb), color='r');
```
